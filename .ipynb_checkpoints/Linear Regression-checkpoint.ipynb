```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Dataset\enhanced_anxiety_dataset cleaned.csv")
```

```python
df
```

```python
features = ['Sleep Hours', 'Caffeine Intake (mg/day)', 'Stress Level (1-10)',
            'Physical Activity (hrs/week)', 'Therapy Sessions (per month)', 'Diet Quality (1-10)']
target = ['Anxiety Level (1-10)']
```

```python
x = df[features]
y = df[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)
print(f"Mean Square Error: {mse:.2f}")
print(f"R-Square: {r2:.2f}")
```

```python
plt.scatter(y_test,y_pred, alpha = 0.7)
plt.xlabel("Anxiety Level (1-10) thực tế")
plt.ylabel("Anxiety Level (1-10) dự đoán")
plt.title("Dự đoán Anxiety Level (1-10) so với thực tế")
plt.plot([0, 10],[0, 10],"r--")
plt.show()
```
